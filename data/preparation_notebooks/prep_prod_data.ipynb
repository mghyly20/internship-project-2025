```python
import pandas as pd
df_production=pd.read_csv('municipalities_production.csv')
df_production
```

```python
alesund_production = df_production[df_production['name'] == 'Ålesund'][[
    'startTime', 'endTime', 'meteringPointTypeCode', 'quantityKwh'
]].copy()

alesund_production
```

```python
alesund_production['meteringPointTypeCode'].unique()
```

```python
#  convertir en datetime avec UTC
alesund_production['startTime'] = pd.to_datetime(alesund_production['startTime'], utc=True)
alesund_production['endTime'] = pd.to_datetime(alesund_production['endTime'], utc=True)

#  supprimer l'information de timezone (résultat : datetime naive)
alesund_production['startTime'] = alesund_production['startTime'].dt.tz_localize(None)
alesund_production['endTime'] = alesund_production['endTime'].dt.tz_localize(None)

```

```python
# Supprimer la colonne 'endTime'
alesund_production.drop(columns=['endTime'], inplace=True)

# Renommer la colonne 'startTime' en 'Date'
alesund_production.rename(columns={'startTime': 'Date'}, inplace=True)
```

```python
alesund_production
```

```python
alesund_production.to_csv('production_alesund_kwh.csv')
```
