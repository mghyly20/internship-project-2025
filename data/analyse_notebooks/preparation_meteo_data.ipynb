La vitesse du vent est donnée par :

$$
 \text{wind} = \sqrt{u_{10}^2 + v_{10}^2} 
$$

```python
df_u = pd.read_csv("u10_alesund.csv")
df_v = pd.read_csv("v10_alesund.csv")

# Assurez-vous qu'ils ont les mêmes lignes (même ordre, date, lat, lon...)
df = df_u.copy()
df["v10"] = df_v["v10"]

# Calcul de la vitesse
df["wind_speed"] = np.sqrt(df["u10"]**2 + df["v10"]**2)

# Sauvegarde
df.to_csv("data_wind_alesund.csv", index=False)


```

```python
df_wind_alesund=pd.read_csv('data_wind_alesund.csv')
df_wind_alesund
```

```python
df_wind_alesund['longitude'].unique()
```

on va concerver juste le point  6.25 pour longitude et 62.5 pour latitude qui represente alesund 

```python
df_wind_alesund = df_wind_alesund[
    (df_wind_alesund['longitude'] == 6.25) &
    (df_wind_alesund['latitude'] == 62.5)
]
df_wind_alesund
```

```python
# Garder uniquement les colonnes souhaitées
df_wind_alesund = df_wind_alesund[['valid_time', 'wind_speed']]
df_wind_alesund
```

```python
max_wind = df_wind_alesund['wind_speed'].max()
min_wind = df_wind_alesund['wind_speed'].min()

print("la vitesse du vent maximale (m/s) :", max_wind)
print("la vitesse du vent minimale (m/s) :", min_wind)
```

```python
df_wind_alesund = df_wind_alesund.rename(columns={"valid_time": "Date"})
```

```python
df_wind_alesund
```

```python
df_wind_alesund.to_csv("wind_speed_alesund.csv", index=False)
```

```python
nan_par_colonne = df_wind_alesund.isna().sum()
print(nan_par_colonne)
```

```python
# Assurez-vous que la colonne est bien en format datetime
df_wind_alesund['Date'] = pd.to_datetime(df_wind_alesund['Date'])

# Calculer la différence entre les lignes
time_diffs = df_wind_alesund['Date'].diff()

# Vérifier si toutes les différences sont exactement de 1 heure
is_one_hour = (time_diffs == pd.Timedelta(hours=1))

# Vérifier s'il y a des valeurs qui ne sont PAS de 1h
if is_one_hour[1:].all():
    print("Toutes les lignes sont espacées d'exactement 1 heure.")
else:
    print("Il y a des lignes qui ne sont pas espacées d'exactement 1 heure.")
```
